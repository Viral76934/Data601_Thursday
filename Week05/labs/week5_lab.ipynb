```python
import seaborn as sns
import pandas as pd
diamonds = sns.load_dataset("diamonds")

diamonds.head()
```

#### 1.

- Plot the scatter plot of price and carat. Add clarity into the plot as hue. 

- Do you see any pattern? Is it parallel to your expectation?

```python
diamonds.info()
diamonds.clarity.unique()
diamonds.color.unique()
```

```python
sns.scatterplot(data = diamonds, x="price", y="carat", hue="clarity", alpha= 0.5)
```

######  => I can be inferred from the plot that diamonds with higher clarity are having carat weight of less that 1.5, and the diamonds with lower clarity seems to have higher carat weight.
###### => However, without having more knowledge of diamonds, it can not be said that clarity and carat weight are inversely proportioned.

#### 2.

- Consider `clarity` and `color` is there any pattern in their heatmap?

```python
sns.displot(data= diamonds, x= "clarity", y= "color")
```

##### => Looking at the heatmap, it can be said that diamonds with lower clarity(VS2, SI1) have color quality above average and is most concentrated in the color 'E'
