##### Creating a Phone_Number column from 'Notes' column

```python
has_phoneno = re.compile('\s*(?:\+?(\d{1,3}))?[-. (]*(\d{3})[-. )]*(\d{3})[-. ]*(\d{4})(?:x(\d{2,6}))?\s*')
```

```python
phno_lst = []
for i in range(0, len(df.loc[:, 'notes'].values)):
    if re.findall(has_phoneno, df.loc[:, 'notes'].values[i]) == []:
        phno_lst.append(has_phoneno.search(df.loc[:, 'notes'].values[i]))
        #print(i, '---', df.loc[:, 'notes'].values[i], '---', re.findall(has_phoneno, df.loc[:, 'notes'].values[i]))
    else:
        phno_lst.append(has_phoneno.search(df.loc[:, 'notes'].values[i]).group())
        print(i, '---', df.loc[:, 'notes'].values[i], '---', has_phoneno.search(df.loc[:, 'notes'].values[i]).group())
```

```python
len(phno_lst)
```

```python
df['Phone_Number'] = phno_lst
```

##### Creating a Start_Date column from 'Notes' column

```python
has_startdate = re.compile('((\d{2,4}|January|February|March|April|May|June|July|August|September|October|November|December)?[-. \/](\d{2,4})?[-. \/]?(\d{2,4})?[-\s\/]?e\s*)')
```

```python
sdt_lst = []
for i in range(0, len(df.loc[:, 'notes'].values)):
    if re.findall(has_startdate, df.loc[:, 'notes'].values[i]) == []:
        continue
        sdt_lst.append(has_startdate.search(df.loc[:, 'notes'].values[i]))
        #print(i, '---', df.loc[:, 'notes'].values[i], '---', re.findall(has_phoneno, df.loc[:, 'notes'].values[i]))
    else:
        sdt_lst.append(has_startdate.search(df.loc[:, 'notes'].values[i]).group())
        print(i, '---', df.loc[:, 'notes'].values[i], '---', has_startdate.search(df.loc[:, 'notes'].values[i]).group(0), '---', has_startdate.search(df.loc[:, 'notes'].values[i]).group(2), '---', has_startdate.search(df.loc[:, 'notes'].values[i]).group(3), '---', has_startdate.search(df.loc[:, 'notes'].values[i]).group(4))
```

```python
len(sdt_lst)
```

```python
df['Start_Date'] = sdt_lst
```

```python
df['Start_Date'] = df['Start_Date'].str.rstrip('e')
df['Start_Date'] = df['Start_Date'].str.rstrip('/')
df['Start_Date'] = df['Start_Date'].str.rstrip('-')
df['Start_Date'] = df['Start_Date'].str.rstrip(' ')
```

```python
sty = re.compile('\d{4}|\d{2}$')
```

```python
start_year = []
for i in range(0, len(df.loc[:, 'Start_Date'].values)):
    if re.findall(sty, df.loc[:, 'Start_Date'].values[i]) == []:
        start_year.append(sty.search(df.loc[:, 'Start_Date'].values[i]))
        print(i, '---', df.loc[:, 'Start_Date'].values[i], '---', re.findall(sty, df.loc[:, 'Start_Date'].values[i]))
    else:
        start_year.append(sty.search(df.loc[:, 'Start_Date'].values[i]).group())
        print(i, '---', df.loc[:, 'Start_Date'].values[i], '---', sty.search(df.loc[:, 'Start_Date'].values[i]).group())
```

```python
for i in range(0, len(start_year)):
    if int(start_year[i]) < 22:
        start_year[i] = '20' + start_year[i]
    elif int(start_year[i]) > 0 and int(start_year[i]) < 100:
        start_year[i] = '19' + start_year[i]
```
