```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get('https://raw.githubusercontent.com/msaricaumbc/DS_Data/master/data.html')
response.ok
```

```python
response.text[:1000]
```

```python
html_doc = response.text
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
t = soup.find_all('div', {"class":"col-md-4 country"})
t
```

```python
t[0]
```

```python
t[0].find('span', {'class':'country-capital', 'class':'country-population', 'class':'country-area'}).text
```

```python
len(t)
```

```python
country_name = []
for i in range(0, len(t)):
    country_name.append(t[i].find('h3').text)
```

```python
country_name
```

```python
country_capital = []
for i in range(0, len(t)):
    country_capital.append(t[i].find('span', {'class':'country-capital'}).text)
```

```python
country_capital
```

```python
country_population = []
for i in range(0, len(t)):
    country_population.append(t[i].find('span', {'class':'country-population'}).text)
```

```python
country_population
```

```python
country_area = []
for i in range(0, len(t)):
    country_area.append(t[i].find('span', {'class':'country-area'}).text)
```

```python
country_area
```

```python
df = pd.DataFrame()
```

```python
df['Country_Name'] = country_name
df['Country_Capital'] = country_capital
df['Country_Population'] = country_population
df['Country_Area'] = country_area
```

```python
df['Country_Population'] = df['Country_Population'].astype('int')
```

```python
df['Country_Area'] = df['Country_Area'].astype('float')
```

```python
df['Country_Name'] = df['Country_Name'].str.strip()
```

```python
df['Country_Capital'] = df['Country_Capital'].str.strip()
```

```python
df
```

```python
df.info()
```

```python
df.to_csv('Country_Data.csv')
```
