```python
final_df = pd.merge(left = df, right = df1, how = 'left', on = 'Country_Name')
```

```python
final_df.info()
```

```python
final_df.fillna(value=0, inplace = True)
```

```python
final_df['2020'] = final_df['2020'].str.replace(',','')
final_df['2021'] = final_df['2021'].str.replace(',','')
final_df['2022'] = final_df['2022'].str.replace(',','')
final_df['2023'] = final_df['2023'].str.replace(',','')
final_df['2024'] = final_df['2024'].str.replace(',','')
final_df['2025'] = final_df['2025'].str.replace(',','')
final_df['2026'] = final_df['2026'].str.replace(',','')
final_df['2027'] = final_df['2027'].str.replace(',','')
```

```python
final_df['2020'] = final_df['2020'].astype(float)
final_df['2021'] = final_df['2021'].astype(float)
final_df['2022'] = final_df['2022'].astype(float)
final_df['2023'] = final_df['2023'].astype(float)
final_df['2024'] = final_df['2024'].astype(float)
final_df['2025'] = final_df['2025'].astype(float)
final_df['2026'] = final_df['2026'].astype(float)
final_df['2027'] = final_df['2027'].astype(float)
```

```python
final_df.info()
```

```python
final_df.shape[0]
```

```python
avg_inc = []
for i in range(0, final_df.shape[0]):
    avg_inc.append(((final_df['2027'][i] - final_df['2022'][i]) / final_df['2022'][i]) * 100.0)
```

```python
avg_inc
```

```python
final_df['Avg_Increase_Percentage'] = avg_inc
```

```python
final_df.sort_values('Avg_Increase_Percentage', ascending = False, inplace = True)
```
